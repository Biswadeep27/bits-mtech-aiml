```python
import networkx as nx

# Create a directed graph
healthcare_knowledge_graph = nx.DiGraph()

# Define entities
entities = [
    ("Gene", {"type": "entity"}),
    ("Protein", {"type": "entity"}),
    ("Disease", {"type": "entity"}),
    ("Drug", {"type": "entity"})
]

healthcare_knowledge_graph.add_nodes_from(entities)

# Define relationships
relationships = [
    ("Gene", "Protein", {"type": "Encodes"}),
    ("Protein", "Disease", {"type": "Associated_with"}),
    ("Drug", "Protein", {"type": "Targets"}),
    ("Drug", "Disease", {"type": "Treats"})
]

healthcare_knowledge_graph.add_edges_from(relationships)

# Example data
gene_attributes = {"Name": "BRCA1", "Function": "DNA repair"}
protein_attributes = {"Name": "BRCA1 protein", "Function": "Tumor suppressor"}
disease_attributes = {"Name": "Breast Cancer", "Symptoms": ["Lump in the breast", "Change in breast shape"]}
drug_attributes = {"Name": "Tamoxifen", "Type": "Selective estrogen receptor modulator"}

# Add attributes to nodes
healthcare_knowledge_graph.nodes["Gene"]["attributes"] = gene_attributes
healthcare_knowledge_graph.nodes["Protein"]["attributes"] = protein_attributes
healthcare_knowledge_graph.nodes["Disease"]["attributes"] = disease_attributes
healthcare_knowledge_graph.nodes["Drug"]["attributes"] = drug_attributes
```

```python
import matplotlib.pyplot as plt

# Draw the knowledge graph
pos = nx.spring_layout(healthcare_knowledge_graph)  # Define layout for better visualization
nx.draw(healthcare_knowledge_graph, pos, with_labels=True, font_weight='bold', node_size=800, node_color='skyblue', font_size=8, arrowsize=15)

# Draw edge labels
edge_labels = {(src, dest): data["type"] for src, dest, data in healthcare_knowledge_graph.edges(data=True)}
nx.draw_networkx_edge_labels(healthcare_knowledge_graph, pos, edge_labels=edge_labels)

# Display the plot
plt.show()
```
